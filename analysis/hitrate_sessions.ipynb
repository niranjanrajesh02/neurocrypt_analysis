```python
def convert_to_str(data_list):
    converted_data = []
    for data_point in data_list:
        converted_data.append(ast.literal_eval(data_point))
    return converted_data
```

```python
def get_blockwise_train_passhr(training_list, pass_specific=False):
    blockwise_train_passhr = []  # 2D list to store each user's blockwise training advantage

    for training_point in training_list:  # Loop through each user's training data
        block_train_passhr = []  # List to store training advantage for each block for a specific user
        for block in range(1, 8):  # Loop through each block (1 to 7)
            for session in training_point:  # Loop through each session in the user's data
                if session['block'] == block:
                    pass_hr = session['passHits'] / (session['passHits'] + session['passMisses'])  # Calculate pass HR
                    block_train_passhr.append(pass_hr)  # Append training advantage to the list
        blockwise_train_passhr.append(block_train_passhr)  # Append the user's blockwise training advantage to the 2D list
    blockwise_train_passhr_array = np.array(blockwise_train_passhr)
    blockwise_avg_train_passhr = np.mean(blockwise_train_passhr_array, axis=0)  # 1D array with 7 values

    return blockwise_avg_train_passhr
```

```python
control_train_passHR = np.mean(get_blockwise_train_passhr(control_train_list))
audio_train_passHR = np.mean(get_blockwise_train_passhr(audio_train_list))
visual_train_passHR = np.mean(get_blockwise_train_passhr(visual_train_list))

print(control_train_passHR)
print(audio_train_passHR)
print(visual_train_passHR)
```

```python
def get_auth_passhr(auth_list, pass_specific=False):
    auth_passhr = []  # list to store each user's blockwise training advantage

    for auth_point in auth_list:  # Loop through each user's training data
        if auth_point['block'] == 1:
            pass_hr = auth_point['passHits'] / (auth_point['passHits'] + auth_point['passMisses'])  # Calculate pass HR
            auth_passhr.append(pass_hr)  # Append training advantage to the list
    auth_passHR = np.mean(auth_passhr)

    return auth_passHR
```

```python
def get_auth_passhr(auth_list, pass_specific=False):
    auth_passhr = []  # list to store each user's blockwise training advantage

    for user_data in auth_list:  # Loop through each user's training data
        for auth_point in user_data: # Loop through each session for the user
            if auth_point['block'] == 1:
                pass_hr = auth_point['passHits'] / (auth_point['passHits'] + auth_point['passMisses'])  # Calculate pass HR
                auth_passhr.append(pass_hr)  # Append training advantage to the list
    auth_passHR = np.mean(auth_passhr)

    return auth_passHR
```

```python
control_auth1_passHR = get_auth_passhr(control_auth1_list)
audio_auth1_passHR = get_auth_passhr(audio_auth1_list)
visual_auth1_passHR = get_auth_passhr(visual_auth1_list)

print(control_auth1_passHR)
print(audio_auth1_passHR)
print(visual_auth1_passHR)
```

```python
def get_auth_perfadv(auth_list, pass_specific=False):
    auth_perf_adv = []  # list to store each user's blockwise training advantage
    perf_adv = 0
    for user_data in auth_list:  # Loop through each user's training data
        for auth_point in user_data: # Loop through each session for the user
            if auth_point['block'] == 1:
                pass_hr = auth_point['passHits'] / (auth_point['passHits'] + auth_point['passMisses'])  # Calculate pass HR
                noise_hr = auth_point['noiseHits'] / (auth_point['noiseHits'] + auth_point['noiseMisses'])  # Calculate noise HR
                perf_adv = pass_hr - noise_hr  # Calculate noise HR
                auth_perf_adv.append(perf_adv)  # Append training advantage to the list
    auth_perfadv = np.mean(auth_perf_adv) # calculate mean of performance advantage for all players of that group

    return auth_perfadv
```

```python
control_auth1_perf_adv = get_auth_perfadv(control_auth1_list)
audio_auth1_perf_adv = get_auth_perfadv(audio_auth1_list)
visual_auth1_perf_adv = get_auth_perfadv(visual_auth1_list)

control_auth2_perf_adv = get_auth_perfadv(control_auth2_list)
audio_auth2_perf_adv = get_auth_perfadv(audio_auth2_list)
visual_auth2_perf_adv = get_auth_perfadv(visual_auth2_list)
```
